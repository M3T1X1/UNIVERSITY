# Content aware image resizing

### Seam carving

Seam carving to algorytm służący do zmiany rozmiarów obrazu z zachowaniem jak największej spójności jego zawartości. Może być przydatny w sytuacjach, gdy tradycyjne metody (skalowanie, wycinanie) nie przynoszą oczekiwanego rezultatu:

[![seam-carving01.png](https://i.postimg.cc/prfFWRxY/seam-carving01.png)](https://postimg.cc/gL0J49xn)


*Seam* (szew) to optymalna, połączona ścieżka pikseli na jednym obrazie, przebiegająca od góry do dołu lub od lewej do prawej. Każdy z pikseli w ścieżce jest połączony z jednym pikselem ze swojego 8-elementowego otoczenia.

<p align="center" width="100%">
    <img width="50%" src="https://i.postimg.cc/cJ6mhgMK/seam-carving02.png">
</p>

Optymalność ścieżki jest definiowana przez funkcję energii obrazu. Poprzez wielokrotne wycinanie lub dodawanie szwów w jednym kierunku możemy zmieniać proporcje obrazu. Zastosowanie tych operatorów w obu kierunkach pozwala dostosować obraz do nowego rozmiaru.

Przejrzyj oryginalny [artykuł](https://www.win.tue.nl/~wstahw/edu/2IV05/seamcarving.pdf).
Zwróć uwagę na *abstract*, opis operatora, przykłady oraz *limitations*. Następnie zapoznaj się z demem, które obrazuje iteracyjne wykonywanie algorytmu: https://www.aryan.app/seam-carving/


Kroki do wykonania:

- Oblicz wielkości gradientu obrazu.

- Znajdź połączoną ścieżkę pikseli, która ma najmniejszą sumę gradientów. Ścieżka jest prawidłowa jeśli jest połączona (sąsiednie punkty na ścieżce są również sąsiednimi pikselami obrazu), zaczyna się w pierwszym rzędzie obrazu i w każdym kroku kontynuuje jeden rząd w dół.

- Usuń piksele na ścieżce z obrazu. W ten sposób otrzymujesz nowy obraz z jedną kolumną mniej.


Implementacja wymaga znalezienia ścieżki w obrazie o najmniejszej sumie gradientów. Jedną z metod, którą można wykorzystać jest programowanie dynamiczne. W przypadku seam carving będzie polegać ono na przechowywaniu skumulowanej sumy dla każdego piksela: energia (gradient) piksela + minimalna wartość skumulowanej energii 3 pikseli nad nim.

Procedurę tą należy powtarzać od drugiego do ostatniego wiersza obrazu. Ostatni krok algorytmu opisuje dobrze sam oryginalny artykuł:

*“At the end of this process, the minimum value of the last row in M will indicate the end of the minimal connected vertical seam. Hence, in the second step we backtrack from this minimum entry on M find the path of the optimal seam (see Figure 1). The definition of M for horizontal seams is similar.”*
