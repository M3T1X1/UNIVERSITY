```python
%reset -f

import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, recall_score


train = pd.read_csv('./dz1_train.csv', sep=';')
test = pd.read_csv('./dz1_test.csv', sep=';')

X_train = train.iloc[:, :-1]
y_train = train.iloc[:, -1]
X_test = test.iloc[:, :-1]
y_test = test.iloc[:, -1]

knn = KNeighborsClassifier(n_neighbors=5, metric='euclidean')

knn.fit(X_train, y_train)

y_pred = knn.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
recall_class_0 = recall_score(y_test, y_pred, pos_label=0)
recall_class_1 = recall_score(y_test, y_pred, pos_label=1)

print(f'Dokładność klasyfikacji: {accuracy:.4f}')
print(f'Recall dla klasy 0: {recall_class_0:.4f}')
print(f'Recall dla klasy 1: {recall_class_1:.4f}')
```

```python
%reset -f

import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, recall_score


train = pd.read_csv('./dz1_train.csv', sep=';')
test = pd.read_csv('./dz1_test.csv', sep=';')

X_train = train.iloc[:, :-1]
y_train = train.iloc[:, -1]
X_test = test.iloc[:, :-1]
y_test = test.iloc[:, -1]

dt = DecisionTreeClassifier(
    criterion='entropy',
    max_depth=6,
    min_impurity_decrease=0.01
)

dt.fit(X_train, y_train)

y_pred = dt.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
recall_class_0 = recall_score(y_test, y_pred, pos_label=0)
recall_class_1 = recall_score(y_test, y_pred, pos_label=1)

print(f'Dokładność klasyfikacji: {accuracy:.4f}')
print(f'Recall dla klasy 0: {recall_class_0:.4f}')
print(f'Recall dla klasy 1: {recall_class_1:.4f}')
```

```python
# Klasyfikator k-NN osiągnął accuracy na poziomie 65,97%,
# a recall 75,5% dla klasy 0 i 55,68% dla klasy 1.

# Klasyfikator drzewa decyzyjnego osiągnął accuracy na poziomie 77,14%,
# a recall 86,0% dla klasy 0 i 67,57% dla klasy 1.

# Drzewo decyzyjne wykazuje lepszą skuteczność.
```

```python
%reset -f

import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier

train = pd.read_csv('./dz1_train.csv', sep=';')
test = pd.read_csv('./dz1_test.csv', sep=';')

X_train = train.iloc[:, :-1]
y_train = train.iloc[:, -1]
X_test = test.iloc[:, :-1]
y_test = test.iloc[:, -1]

knn = KNeighborsClassifier(n_neighbors=5, metric='euclidean')
knn.fit(X_train, y_train)
y_pred_knn = knn.predict(X_test)

dt = DecisionTreeClassifier(
    criterion='entropy',
    max_depth=6,
    min_impurity_decrease=0.01
)
dt.fit(X_train, y_train)
y_pred_dt = dt.predict(X_test)

#  - 0 -> (TN, TP)
#  - 1 -> (FN: prawda=1, pred=0)
#  - 5 -> (FP: prawda=0, pred=1)

def oblicz_koszt(y_true, y_pred):
    koszt = 0
    for yt, yp in zip(y_true, y_pred):
        if yt == 1 and yp == 0:      # False Negative
            koszt += 1
        elif yt == 0 and yp == 1:    # False Positive
            koszt += 5
    return koszt

koszt_knn = oblicz_koszt(y_test, y_pred_knn)
koszt_dt = oblicz_koszt(y_test, y_pred_dt)

print(f'Koszt decyzji k-NN: {koszt_knn}')
print(f'Koszt decyzji drzewa: {koszt_dt}')
```

```python
# Klasyfikator k-NN osiągnął koszt 327,
# natomiast drzewo decyzyjne osiągnęło koszt na poziomie 189.

# Także w tym wypadku drzewo okazuje się lepszym klasyfikatorem.
```
